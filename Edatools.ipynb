```python
plt.figure(figsize=(10,5))  # Sets the figure size to 10 inches wide and 5 inches tall
plt.plot(x, y,              # Plots x vs y
         color='red',       # Line color is red
         marker='o',        # Adds circular markers at each data point
         linewidth=5,       # Line width is 5
         markersize=5)      # Marker size is 5
plt.xlabel('x-axis')        # Label for the x-axis
plt.ylabel('y-axis')        # Label for the y-axis
plt.title('Line Plot')      # Title of the plot
plt.show()                  # Displays the plot
```

Example: Market share of companies

```python
# pie char
#Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]

# Create a simple pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.title("Programming Language Popularity")
plt.axis('equal')  # Keeps the pie chart round

# Show the plot
plt.show()
```

#📊 When to Use Different Matplotlib Plots: Scenarios & Examples

1️⃣ Line Plot

Purpose:
To show trends over time or ordered data (how something changes).

Use when:

You have data points collected over intervals (days, months, years).

You want to show growth, decline, or patterns.

Example scenarios:

Stock prices over days

Temperature changes during a week

Sales growth over quarters

2️⃣ Bar Plot

Purpose:
To compare categories with respect to a value.

Use when:

You have discrete categories (e.g., products, cities, brands).

You want to show how one category stacks up against others.

Example scenarios:

Number of books sold by genre

Population of different countries

Monthly sales revenue comparison


3️⃣ Histogram
Purpose:
To show the distribution of numerical data — how data is spread over ranges (bins).

Use when:

You want to understand the shape of data (normal, skewed, uniform).

You want to see frequency or count in intervals.

Example scenarios:

Ages of people in a survey

Test scores distribution in a class

Distribution of salaries in a company


4️⃣ Pie Chart
Purpose:
To show proportions or percentages of a whole.

Use when:

You want to visualize parts of a total.

You have a small number of categories (ideally less than 6).

Example scenarios:

Market share of companies

Budget allocation across departments

Survey results for favorite colors


5️⃣ Scatter Plot (Bonus!)
Purpose:
To show relationship or correlation between two variables.

Use when:

You want to see if two variables relate (positively, negatively, or no relation).

You want to identify clusters or outliers.

Example scenarios:

Height vs weight of people

Advertising budget vs sales revenue

Hours studied vs exam score



# An introduction to seaborn

Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.
