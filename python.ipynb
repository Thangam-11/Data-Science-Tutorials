# Python ‚Äì Modify Strings

In Python, strings are immutable. That means we can‚Äôt change the original string directly, but we can create a new modified version using string methods.


Python has a set of built-in methods that you can use on strings.

1. The upper() method returns the string in upper case
2. The lower() method returns the string in lower case
3. The strip() method removes any whitespace from the beginning or the end
4. the split() Splits string into list (by spaces or symbol)
5. the replace() Replaces part of the string

```python
text = "  Welcome to Python class  "

print(text.strip())            # Removes leading/trailing spaces
print(text.upper())            # Converts all to uppercase
print(text.lower())            # Converts all to lowercase
print(text.replace("Python", "Data Science"))  # Replace a word
print(text.split())            # Split into words by space
```

# definitions for each concept:

**Ordered :** As of Python 3.7, dictionaries keep the order of items in the order they were added. In earlier versions (3.6 and below), dictionaries did not guarantee any specific order.

 **Unordered :** In versions before Python 3.7, dictionaries did not have a fixed order for their items.

 **Changeable**: You can modify a dictionary by adding, changing, or removing items after it has been created.

 **Duplicates Not Allowed**: A dictionary cannot have two items with the same key. Each key must be unique.



```python
# Ordered (Python 3.7 and above):
my_dict = {'a': 1, 'b': 2, 'c': 3}
for key, value in my_dict.items():
    print(key, value)
```

```python
# Changeable (Mutable)

my_dict = {'a': 1, 'b': 2}
# Adding an item
my_dict['c'] = 3
# Modifying an existing item
my_dict['b'] = 10
# Removing an item
del my_dict['a']
print(my_dict)
```

```python
#üîπ 2. update(iterable)
#Adds multiple elements from an iterable (like list, tuple, etc.).
s.update([4, 5])
print(s)  # {1, 2, 3, 4, 5}
```

```python
# üîπ 3. remove(elem)
# Removes an element. Raises KeyError if the element is not found.

s.remove(5)
# s.remove(10)  # KeyError if 10 not in set
```

```python
##üîπ 4. discard(elem)
#Removes an element if it exists. No error if not found.

s.discard(10)  # No error
```

```python
#üîπ 5. pop()
#Removes and returns an arbitrary (random) element.

val = s.pop()
print(val)
```

```python
 #6. clear()
#Removes all elements from the set.

s.clear()
print(s)  # Outpu
```

```python
#  7. copy()
#Returns a shallow copy of the set.

a = {1, 2}
b = a.copy()
print(b)
```

```python
#üß† Set Operations Methods
#üîπ 8. union(other_set) or |
#Combines sets.

a = {1, 2}
b = {2, 3}
print(a.union(b))  # {1, 2, 3}
```

```python
#  9. intersection(other_set) or &

#Gets common elements.

print(a.intersection(b))


```

```python
#  10. difference(other_set) or -
# Gets elements in a but not in b.

print(a.difference(b))  # {1}
```

```python

#üîπ 11. symmetric_difference(other_set) or ^
#Gets elements in either set but not both.

print(a.symmetric_difference(b))  # {1, 3}
```

# if - elif - else (Decision Making)

Think of it like making choices:

üî∏ ‚ÄúIf this is true, do this. If not, try the next condition. If nothing is true, do something else.‚Äù



if else operations.jpg

```python
# Example if statement

a = 33
b = 200
if b > a:
  print("b is greater than a")
```

```python
# Elif  example

#The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition".

a = 33
b = 33
if b > a:

   print("b is greater than a")
elif a == b:
  print("a and b are equal")

# In this example a is equal to b, so the first condition is not true, but the elif condition is true, so we print to screen that "a and b are equal".
```

```python
# Else staement example

# The else keyword catches anything which isn't caught by the preceding conditions.

a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

# Python For Loops

What is a loop in Python?

A loop is used to repeat a block of code multiple times.

üìå Example in real life:
If you want to say "Hello" 5 times, you don't write it 5 times ‚Äî you use a loop!

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python

# A for loop is used to repeat something for each item in a list, string, or range of numbers.

for i in range(1, 6):
    print(i)
```

```python
# Example 2: Loop through a list

fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

```python
#2. Looping Through a String
#üìò You can also loop through each letter in a word.


for x in "banana":
    print(x)
```

```python
# 3. break Statement
#üìò Used to stop the loop early.

fruits = ["apple", "banana", "cherry"]

for x in fruits:
    print(x)
    if x == "banana":
        break

```

```python
for x in fruits:
    if x == "banana":
        break
    print(x)
```

```python
# 4. continue Statement
# Used to skip one item in the loop.

for x in fruits:
    if x == "banana":
        continue
    print(x)
```

```python
 #5. range() Function
# range(n) gives numbers from 0 to (n-1)
for x in range(6):
    print(x)
```

```python
#  You can also set start and stop:

for x in range(2, 6):
    print(x)
```

```python
# You can also set step size:

for x in range(2, 10, 2):
    print(x)
```

```python

# 6. else with for loop
# else will run after the loop finishes.

for x in range(3):
    print(x)
else:
    print("Loop ended!")
```

```python

# if we use break, else won‚Äôt run:

for x in range(6):
    if x == 3:
        break
    print(x)
else:
    print("Loop ended!")  # ‚Üê This won't run
```

#The while Loop

With the while loop we can execute a set of statements as long as a condition is true.

üîÅ Syntax:

while condition:
    # do something

```python
i = 1

while i <= 5:
    print(i)
    i = i + 1
```

```python

# Using break in a while loop

i = 1

while i <= 5:
    print(i)
    if i == 3:
        break
    i += 1
```

# Warning: Infinite Loop
Make sure the condition becomes False at some point, or it will run forever

while True:
    print("I will run forever!")  # press stop to break it!

```python
n = int(input())

arr = list(map(int, input().split()))

for i in range(n):
    for j in range(0, n - i - 1):  # Reduce range each pass
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print(arr)
```
